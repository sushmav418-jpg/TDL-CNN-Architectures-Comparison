```python
import pandas as pd
import matplotlib.pyplot as plt

# Example: manually create dataframe (safe)
data = {
    "Model": ["LeNet", "AlexNet", "ResNet"],
    "MNIST": [99.0, 98.2, 99.3],
    "CIFAR-10": [65.4, 74.8, 82.1]
}

df = pd.DataFrame(data)
df.set_index("Model").plot(kind="bar")

plt.ylabel("Accuracy (%)")
plt.title("Model Accuracy Comparison")
plt.show()
```

```python
time_data = {
    "Model": ["LeNet", "AlexNet", "ResNet"],
    "MNIST Time": [20, 65, 90],
    "CIFAR-10 Time": [45, 180, 260]
}

df_time = pd.DataFrame(time_data)
df_time.set_index("Model").plot(kind="bar")

plt.ylabel("Training Time (seconds)")
plt.title("Training Time Comparison")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Model": ["LeNet", "AlexNet", "ResNet"],
    "MNIST": [98.68, 99.07, 99.07],
    "CIFAR-10": [55.03, 75.33, 41.53]
}

df = pd.DataFrame(data).set_index("Model")
df.plot(kind="bar")

plt.ylabel("Accuracy (%)")
plt.title("Accuracy Comparison of CNN Models")
plt.show()
```

```python
time_data = {
    "Model": ["LeNet", "AlexNet", "ResNet"],
    "MNIST": [22.14, 52.52, 119.25],
    "CIFAR-10": [22.19, 53.33, 138.37]
}

df_time = pd.DataFrame(time_data).set_index("Model")
df_time.plot(kind="bar")

plt.ylabel("Training Time (seconds)")
plt.title("Training Time Comparison of CNN Models")
plt.show()
```
